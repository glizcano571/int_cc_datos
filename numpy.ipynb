# MATPLOTLIB LIBRERIA PARA GRAFICAR

```python
tiempo = np.linspace(0, 60, 100) #genera 100 numeros en un rango de 1 a 10
crecimiento = 5*tiempo-10*np.random.rand(100) #crece 5 veces el tiempo y le agregue una variable de numero aleatorio
```

```python
!pip install matplotlib #instalar libreria
```

```python
import matplotlib.pyplot  as plt
```

```python
plt.plot(tiempo, crecimiento, '<', color = 'black') #aqui lo que hace es relacionar las dos variables anteriores con los datos que hay mediante un grafico
# parte horizontal tiempo, parte vertical crecimiento
plt.xlabel('Tiempo(s)') # en el eje x ponemos ahi el nombre Tiempo
plt.ylabel('Crecimiento(cm)') #en el eje y ponemos ahi el nombre crecimiento
plt.title('Crecimiento vs Tiempo') #titulo de la grafica
plt.grid() #grilla de la grafica
plt.legend(['Arbol ']) # aqui es para identicar cual crecimiento es de cada arbol segun la grafica
```

```python
tiempo = np.linspace(0, 60, 100) #genera 100 numeros del 1 al 10
crecimiento1 = 5*tiempo-10*np.random.rand(100) #crece 5 veces el tiempo, tiene agregada una variable de numeros aleatorios
crecimiento2 = 5.8*tiempo-12*np.random.rand(100)
```

```python
plt.plot(tiempo, crecimiento1,  '<', color = 'yellow' ) #LA IDEA ES MIRAR LA DIFERENCIA ENTRE AMBOS CRECIMIENTOS DE LOS ARBOLES 'crecimiento 1 y crecimiento 2'
plt.plot(tiempo, crecimiento2, 'o', color = 'black')
plt.xlabel('Tiempo(s)')
plt.ylabel('Crecimiento(cm)')
plt.title('Crecimiento vs Tiempo') #tìtulo de la grafica
plt.grid() #grilla de la grafica
plt.legend(['Arbol 1','Arbol 2']) #identicar el crecimiento es de cada arbol segùn la grafica
```
